```
import numpy as np
import pandas as pd
from sklearn.metrics import *
from sklearn.model_selection import *

from tqdm import tqdm
import warnings
warnings.simplefilter('ignore')
import gc
from scipy.special import softmax

from simpletransformers.classification import MultiLabelClassificationModel
from sklearn.metrics import mean_squared_error as mse
```

```
train_df = pd.read_csv('./sample_data/Train.csv')
test = pd.read_csv('./sample_data/Test.csv')
train = pd.get_dummies(train_df['label'])
train['ID'] = train_df['ID']
train['text'] = train_df['text']
```

```
df = train
```

```
df.head()
```

```
df['labels'] = list(zip(df.Alcohol.tolist(), df.Depression.tolist(), df.Drugs.tolist(), df.Suicide.tolist()))
df['text'] = df['text'].apply(lambda x: x.replace('\n', ' '))
df.head()
```

```
#MODEL USED roberta-base
from sklearn.model_selection import train_test_split
from simpletransformers.classification import MultiLabelClassificationModel

train_df, eval_df = train_test_split(df, test_size=0.2)
model = MultiLabelClassificationModel('roberta', 'roberta-base',use_cuda=False, num_labels=4, args={'train_batch_size':2, 'gradient_accumulation_steps':16, 'learning_rate': 3e-5, 'num_train_epochs': 1, 'max_seq_length': 512})
# Train the model
model.train_model(train_df)
result, model_outputs, wrong_predictions = model.eval_model(eval_df)


```

```
true = eval_df['labels'].tolist()
tr = []
for i in true:
  i = list(i)
  tr.append(i)
from sklearn.metrics import log_loss
ev = log_loss(true ,model_outputs)
print('Log Loss:' , ev)
```

```
#MODEL USED roberta-large
from sklearn.model_selection import train_test_split
from simpletransformers.classification import MultiLabelClassificationModel

train_df, eval_df = train_test_split(df, test_size=0.2)
model_roberta_large = MultiLabelClassificationModel('roberta', 'roberta-large',use_cuda=False, num_labels=4, args={'train_batch_size':2, 'gradient_accumulation_steps':16, 'learning_rate': 3e-5, 'num_train_epochs': 1, 'max_seq_length': 512})
# Train the model
model_roberta_large.train_model(train_df, output_dir = './output_rob_l')
result_roberta_large, model_outputs_roberta_large, wrong_predictions_roberta_large = model_roberta_large.eval_model(eval_df)

```
